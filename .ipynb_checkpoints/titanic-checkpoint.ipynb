# Data Dictionary

- PassengerId : 탑승객의 고유 아이디
- Survival : 생존여부(0: 사망, 1: 생존)
- Pclass : 등실의 등급(1: 1등급, 2: 2등급, 3: 3등급)
- Name : 이름
- Sex : 성별
- Age : 나이
- Sibsp : 함께 탑승한 형제자매, 아내 남편의 수
- Parch: 함께 탑승한 부모, 자식의 수
- Ticket: 티켓번호
- Fare: 티켓의 요금
- Cabin: 객실번호
- Embarked: 배에 탑승한 위치(C = Cherbourg, Q = Queenstown, S = Southampton)

### 데이터 전처리
- 결측치 처리
- feature scaling
- 데이터 분석

```python
# embarked 결측치 대체 (최빈값)
train_data['Embarked'] = train_data['Embarked'].fillna('S')
train_data.isna().sum()
```

```python
train_data.corr()
```

```python
train_data_Age_Pclass = train_data.groupby(train_data["Pclass"]).mean()
train_data_Age_Pclass['Age']
```

```python
train_data.loc[(train_data.Age.isna())&(train_data.Pclass ==1),'Age']= train_data_Age_Pclass['Age'][1]
train_data.loc[(train_data.Age.isna())&(train_data.Pclass ==2),'Age']= train_data_Age_Pclass['Age'][2]
train_data.loc[(train_data.Age.isna())&(train_data.Pclass ==3),'Age']= train_data_Age_Pclass['Age'][3]
#train_data['Age'] = train_data[(train_data['Age'].isna())&(train_data['Pclass']==1)]['Age'].fillna(train_data_Age_Pclass['Age'][1])
#train_data['Age'] = train_data[(train_data['Age'].isna())&(train_data['Pclass']==2)]['Age'].fillna(train_data_Age_Pclass['Age'][2])
#train_data['Age'] = train_data[(train_data['Age'].isna())&(train_data['Pclass']==3)]['Age'].fillna(train_data_Age_Pclass['Age'][3])
#train_data[(train_data['Age'].isna())&(train_data['Pclass']==2)]['Age'] = train_data_Age_Pclass['Age'][2]
#train_data[(train_data['Age'].isna())&(train_data['Pclass']==3)]['Age'] = train_data_Age_Pclass['Age'][3]
#train_data[(train_data['Age'].isna())&(train_data['Pclass']==2)]['Age'] = train_data_Age_Pclass['Age'][2]
#train_data[(train_data['Age'].isna())&(train_data['Pclass']==3)]['Age'] = train_data_Age_Pclass['Age'][3]
train_data['Age'].isna().sum()
```
